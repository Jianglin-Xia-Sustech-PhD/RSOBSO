Figure: State Switching Process

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 100, 1000)
```

3 constant piecewise functions

```python
def piece_1(x):
    if x>=0 and x<=30:
        return 30
    else:
        return 0

def piece_2(x):
    if x>=30 and x<=50:
        return 50
    else:
        return 0

def piece_3(x):
    if x>=50 and x<=80:
        return 40
    else:
        return 0

def piece_4(x):
    if x>=80 and x<=100:
        return 50
    else:
        return 0   
```

```python
a1 = np.linspace(0, 30, 1000)
m1 = [piece_1(i) for i in a1]

a2 = np.linspace(30, 50, 1000)
m2 = [piece_2(i) for i in a2]

a3 = np.linspace(50, 80, 1000)
m3 = [piece_3(i) for i in a3]

a4 = np.linspace(80, 100, 1000)
m4 = [piece_4(i) for i in a4]
```

Add random scatter points

```python
np.random.seed(0)
x1 = np.linspace(0, 30, 30)
n1 = np.random.normal(30, 1, 30)
x2 = np.linspace(30, 50, 20)
n2 = np.random.normal(50, 1, 20)
x3 = np.linspace(50, 80, 30)
n3 = np.random.normal(40, 1, 30)
x4 = np.linspace(80, 100, 20)
n4 = np.random.normal(50, 1, 20)
```

```python
plt.figure(figsize=(16, 11))

# 设置全局字体大小（标题、标签、图例、刻度等）
plt.rcParams.update({
    'font.size': 21,          # 控制所有文本字体大小
    'axes.labelsize': 23,     # x轴、y轴标签
    'xtick.labelsize': 24,    # x轴刻度
    'ytick.labelsize': 24,    # y轴刻度
    'legend.fontsize': 31,    # 图例文字
    'figure.titlesize': 21    # 图标题
})

plt.plot(a1, m1, 'red')
plt.plot(a2, m2, 'blue', linestyle = "dashdot")
plt.plot(a3, m3, 'green', linestyle = "dashed")
plt.plot(a4, m4, 'blue', linestyle = "dashdot")
plt.scatter(x1, n1, s = 3, c = "black")
plt.scatter(x2, n2, s = 3, c = "black")
plt.scatter(x3, n3, s = 3, c = "black")
plt.scatter(x4, n4, s = 3, c = "black")
plt.xlabel('t', fontsize = 48)
plt.ylabel(r'$\xi$', rotation=0, labelpad=16, fontsize = 48)
#plt.title('Regime Switching Process')
plt.legend(labels=['Regime 1','Regime 2', 'Regime 3'], loc='best')
plt.vlines(30, 0, 30, colors = 'black', linestyles = 'dashed')
plt.vlines(50, 0, 40, colors = 'black', linestyles = 'dashed')
plt.vlines(80, 0, 40, colors = 'black', linestyles = 'dashed')
plt.xticks([30, 50, 80], labels = [r'$t_1$', r'$t_2$', r'$t_3$'], fontsize = 48)
plt.yticks([], labels = [])
plt.ylim(25, 60)
plt.xlim(0, 100)
plt.arrow(30, 30, 0, 19, head_width = 2, head_length = 1, width = 0.1, color = 'black')
plt.arrow(50, 50, 0, -9, head_width = 2, head_length = 1, width = 0.1, color = 'black')
plt.arrow(80, 40, 0, 9, head_width = 2, head_length = 1, width = 0.1, color = 'black')
plt.text(22, 40, 'switching', weight = "bold", fontsize = 31)
plt.text(42, 45, 'switching', weight = "bold", fontsize = 31)
plt.text(72, 45, 'switching', weight = "bold", fontsize = 31)
plt.savefig('regimes_Aug12.pdf', bbox_inches = 'tight')
plt.show()
```
